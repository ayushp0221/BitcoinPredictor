## Modeling

To create an effective prediction model, we have chosen to use the 2021 Bitcoin data exclusively. This decision is based on the significant increase in Bitcoin's popularity and value during this period. Before 2021, Bitcoin's price was low and highly unpredictable. Using this data could make our model less accurate because it might include a lot of random fluctuations that don't reflect recent trends. So, we're focusing on 2021 data because that's when Bitcoin became more popular and its value increased significantly. By focusing solely on the 2021 data, we aim to capture the recent trends and patterns that are more relevant to current market dynamics. This approach allows us to build a prediction model that is better aligned with the recent behavior of Bitcoin prices, thus enhancing its accuracy and reliability for forecasting future price movements.

#### Without Scaling the data

```python
def create_sequences(seq, steps):
    X, y = [], []
    for i in range(len(seq) - steps):
        X.append(seq[i:i+steps])
        y.append(seq[i+steps])
    return np.array(X), np.array(y)
```

#### With Scaling the Data

```python
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled_data = scaler.fit_transform(np.array(df_train).reshape(-1,1))
test_scaled_data = scaler.transform(np.array(df_test).reshape(-1,1))
```

```python
 model.fit(trainX, trainY, epochs=10, batch_size=64, validation_split=0.35)
```

```python
lstm_predictions = model.predict(testX)
```

# Bidirectional LSTM

```python
model = Sequential()

model.add(Bidirectional(LSTM(50, return_sequences=True, input_shape=(5, 1), activation='tanh')))
model.add(Dropout(0.2))

model.add(Bidirectional(LSTM(25, return_sequences=True, activation='tanh')))
model.add(Dropout(0.2))

model.add(Bidirectional(LSTM(16, return_sequences=True, activation='tanh')))
model.add(Dropout(0.2))

model.add(Bidirectional(LSTM(10, return_sequences=False, activation='tanh')))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(trainX, trainY, epochs=10, batch_size=64, validation_split=0.35)
```

```python
bi_prediction = model.predict(testX)
```

```python
testY = scaler.inverse_transform(testY)
predictions = scaler.inverse_transform(bi_prediction)

testY, predictions
```

```python
mse = mean_squared_error(testY, predictions)
mae = mean_absolute_error(testY, predictions)
rmse = math.sqrt(mse)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```
